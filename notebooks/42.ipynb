```python
with open('0042_words.txt') as file:
    words = file.read()
```

```python
data =[word.strip('"') for word in words.split(',')]
```

```python
def char_index(s):
    return ord(s) - ord('A') + 1
```

```python
def word_value(s):
    tot = 0
    for word in s:
        tot += char_index(word)
    return tot
```

```python
largest = 30 * 24
tri = set([int(i*(i+1)/2) for i in range(1,largest)])
```

```python
count = 0
for word in data:
    if word_value(word) in tri:
        print(word, word_value(word))
        count+=1
print(count)
```
