```python
import pandas as pd
import matplotlib.pyplot as plt
import math

def for_x(i):
    tot = 0
    for j in range(i):
        tot += 9**5
    return tot
lim = 7
data = {'y':[math.log10(for_x(x)) - x for x in range(1,lim)], 'x':[y for y in range(1,lim)]}
df = pd.DataFrame(data)
```

```python
df.plot(x='x',y='y')
```
