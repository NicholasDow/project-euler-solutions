```python
import math
import time
import matplotlib.pyplot as plt

# Original version
def original_sum_primes_below(largest):
    valid = [True for i in range(2, largest)]
    atn = [i for i in range(2, largest)]
    for i, v in enumerate(atn.copy()):
        if valid[i]:
            for j in range(i+1, largest - 2):
                if valid[j] and atn[j] % v == 0:
                    atn[j] = 0
                    valid[j] = False
    return sum(atn)

# Optimized version
def optimized_sum_primes_below(limit):
    if limit <= 2:
        return 0
    
    sieve = bytearray([1]) * (limit // 2)
    sum_primes = 2  # Start with 2, the only even prime
    
    for i in range(3, int(limit**0.5) + 1, 2):
        if sieve[i // 2]:
            sieve[i*i//2::i] = bytearray((limit-i*i-1)//(2*i)+1)
    
    sum_primes += sum(i*2 + 1 for i in range(1, limit//2) if sieve[i])
    
    return sum_primes
optimized_sum_primes_below(2_000_000)
```
