```python
def sieve(n):
    primes = [True]*n
    for i in range(2, int(n**(0.5)) + 1):
        if primes[i]:
            primes[i*i::i] = [False]*len(range(i*i,len(primes),i))
    
    return [i for i in range(n) if primes[i]]
```

```python
primes = sieve(1000)
```

```python
def getCycle(n):
    numerator = 10
    digits = []
    remainders = []
    cur = numerator
    while True:
        digit = cur // n
        cur = cur % n
        print(cur)
        if cur in remainders:
            print('breaking')
            break
        digits.append(digit)
        remainders.append(cur)
        print("data")
        print(digits)
        print(remainders)
        cur *= 10
        if cur == 0:
            return 0
        while cur < n:
            print(cur, n, 'increasing num')
            cur *= 10
            digits.append(0)
    while remainders[0] != cur:
        print('trap')
        remainders.pop(0)
    return len(remainders)
```

```python
getCycle(987)
```

```python
def solution_long_division() -> int:
    cycle_lengths = {denominator: getCycle(denominator) for denominator in range(1, 1000)}
    return sorted(cycle_lengths.items(), key=lambda item: item[1])[-1][0]
```
