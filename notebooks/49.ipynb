```python
valid = [True] * 10
permutation = []
permutations = []

def search():
    # print('running')
    # print('permutation', permutation)
    # print('valid', valid)
    if len(permutation) == 4:
        print('yay we got a combo')
        permutations.append(permutation)
    
    for i in range(10):
        print('index',i)
        if not valid[i]:
            print('stopping')
            continue

        permutation.append(i)
        valid[i] = False
        search()
        print(permutation)
        permutation.pop()
```
